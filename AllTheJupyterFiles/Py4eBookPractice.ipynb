```python
def chop(l):
    del l[:-len(l)+1]
    del l[-1:]
    return 
def middle(l):
    n_list = l[1:-1]
    return n_list
l = list()
while True:
    w = input('Enter the list elements: ')
    if w == 'done' : break
    else:
        l.append(w)
i = input('Enter path: ')
if i == 'chop':
    chop(l)
    print(l)
elif i == 'middle':
    new_list = middle(l)
    print(new_list)
else:
    print('Invalid path')
```

## Tuples (Chapter 10)

### Exercise 1

```python
fhandle = input('Enter thr file name:')
try:
    handle = open(fhandle)
except:
    print("File doesn't exist or not in the same directory...")
    quit()
emails = dict()
l=list()
for line in handle:
    line= line.rstrip()
    if line.startswith('From '):
        words = line.split()
        email=words[1]
        emails[email] = emails.get(email,0)+1
    else:
        continue
for a,b in emails.items():
    l.append((b,a))
l.sort(reverse=True)
for value,key in l[:1]:
    print(key,value)
```

```python
fhandle = input('Enter thr file name:')
try:
    handle = open(fhandle)
except:
    print("File doesn't exist or not in the same directory...")
    quit()
hours = dict()
l =list()
for line in handle:
    line= line.rstrip()
    if line.startswith('From '):
        words = line.split()
        w=words[5]
        h = w.find(':')
        hour = w[h-2:h]
        hours[hour] = hours.get(hour,0)+1
    else:
        continue
for a,b in hours.items():
    l.append((a,b))
l.sort()
for a,b in l:
    print(a,b)
```

### Exercise 3

```python
import string
fhandle = input('Enter thr file name:')
try:
    handle = open(fhandle)
except:
    print("File doesn't exist or not in the same directory...")
    quit()
alpha= dict()
l =list()
for line in handle:
    line= line.rstrip()
    line = line.translate(str.maketrans('','',string.punctuation))
    line = line.lower()
    words = line.split()
    for word in words:
        for letter in word:
            try:
                int(letter)
            except:
                alpha[letter] = alpha.get(letter,0)+1
    else:
        continue
for a,b in alpha.items():
    l.append((b,a))
l.sort(reverse=True)
for b,a in l:
    print(a,b)
```

## Regular Expressions (Chapter 11)

### Exercise 1

```python
import re
handle = open('mbox.txt')
count = 0
re_ex= input('Enter a regular expression:')
for line in handle:
    x= re.search(re_ex,line)
    if x is not None:
        count +=1
print ("mbox.txt had "+str(count)+" lines that matched "+re_ex)
```

### Exercise 2

```python
import re
fhandle=input('Enter file:')
handle = open(fhandle)
x = list()
for line in handle:
    a=re.findall('^N.*: ([0-9]+)',line)
    if len(a) != 0:
        x.append(a[0])
    else:
        continue
b = [int(i) for i in x]
avg = int(sum(b)/len(b))
print(avg)   
```
